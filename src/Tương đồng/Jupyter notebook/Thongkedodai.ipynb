```python
import os
import librosa
import matplotlib.pyplot as plt

input_folder = r"C:\Users\Asus\OneDrive - m2xk\Desktop\Multimedia-DB\src\AudioFile"

lengths = []

for filename in os.listdir(input_folder):
    if filename.endswith(".flac"):  # Hoặc định dạng file khác
        file_path = os.path.join(input_folder, filename)
        y, sr = librosa.load(file_path, sr=None)
        lengths.append(len(y) / sr * 1000)  # Chuyển đổi độ dài từ giây sang mili giây

plt.hist(lengths, bins=20)
plt.xlabel('Length (ms)')
plt.ylabel('Number of Files')
plt.title('Distribution of Audio File Lengths')
plt.show()

# in ra độ dài của file ngắn nhất và dài nhất 
if lengths:
    min_length = min(lengths)
    max_length = max(lengths)
    
    print(f"Shortest file length: {min_length:.2f} ms")
    print(f"Longest file length: {max_length:.2f} ms")
```
