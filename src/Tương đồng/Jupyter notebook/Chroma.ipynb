```python
# Chroma thể hiện cao độ (pitch) của âm thanh, có 12 cao độ khác nhau, mỗi cao độ biểu diễn bằng 1 màu cụ thể
# Chroma thể hiện mức độ xuất hiện của mỗi cao độ trong 1 thời gian ngắn của tín hiệu âm thanh
# Các cao độ tạo nên 1 đường thẳng uốn lượn, tại mỗi điểm có tần số Hz khác nhau
```

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt

import os
os.chdir(r"C:\Users\Asus\OneDrive - m2xk\Desktop\Multimedia-DB\src\AudioFileCollection")
```

```python
# Đường dẫn đến file âm thanh
audio_path = "file1-001.wav"

# Load file âm thanh và tần số lấy mẫu
y, sr = librosa.load(audio_path)

# Trích xuất đặc trưng Chroma
chroma = librosa.feature.chroma_stft(y=y, sr=sr)
print("Chroma feature:")
print(chroma)

# Hiển thị biểu diễn chroma
plt.figure(figsize=(10, 4))
librosa.display.specshow(chroma, y_axis='chroma', x_axis='time')
plt.colorbar()
plt.title('Chromagram')
plt.xlabel('Time')
plt.ylabel('Chroma')
plt.show()
```
