```python
import matplotlib.pyplot as plt
from scipy.io import wavfile
from python_speech_features import logfbank

import os
os.chdir(r"C:\Users\Asus\OneDrive - m2xk\Desktop\Multimedia-DB\src\AudioFileCollection")
```

```python
# Đường dẫn đến file âm thanh WAV
audio_path = "file1-001.wav"

# Đọc file âm thanh và tính Filter Bank
frequency_sampling, audio_signal = wavfile.read(audio_path)
audio_signal = audio_signal[:15000]  # Chọn một phần của tín hiệu âm thanh (nếu cần)
filterbank_features = logfbank(audio_signal, frequency_sampling)

# Hiển thị đặc trưng Filter Bank
print('\nFilter bank:\nNumber of windows =', filterbank_features.shape[0])
print('Length of each feature =', filterbank_features.shape[1])

filterbank_features = filterbank_features.T
plt.matshow(filterbank_features)
plt.title('Filter bank')
plt.show()
```
