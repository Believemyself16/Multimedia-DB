```python
def Find_Max(arr, f, l):
    max = arr[f]
    
    for i in range(f, l):
        if max < arr[i]:
            max = arr[i]
            
    return max

def Find_Min(arr, f, l):
    min = arr[f]
    
    for i in range(f, l):
        if min > arr[i]:
            min = arr[i]
            
    return min

def KhoangLang(arr, f, l):
    max = Find_Max(arr, f, l)
    min = Find_Min(arr, f, l)
    tb = (max + min) / 2 
    nguong = tb * 0.3
    dem = 0
    
    for i in range(f, l):
        if arr[i] > nguong:
            dem = dem + 1

    return 100 - (dem / (l - f)) * 100 
```

```python
plt.figure(figsize=(5, 5))
path = "file1-001.wav"

y, sr = librosa.load(path)
result = []
mfcc = librosa.feature.mfcc(y=y, sr=sr) #mel frequency

print(mfcc.shape)
print("Khoảng lặng theo 1s:")
for i in range(0, 6): #trong khoảng thời gian 6s
    print("thời gian: ", i, " = ",KhoangLang(y, i*sr, (i+1)*sr))
 
librosa.display.specshow(mfcc, x_axis='time')
plt.colorbar()
plt.title('Biểu đồ MFCC')
plt.xlabel('Thời gian')
plt.ylabel('MFCC Coefficients')
plt.tight_layout()
plt.show()
```
